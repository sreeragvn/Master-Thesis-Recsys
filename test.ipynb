```python
import matplotlib.pyplot as plt

# Sample data: Assuming 'x' represents hours in a driving session
x = [1, 2, 3, 4, 5]  # Example: 1-5 hours
y = [2, 3, 5, 7, 11]  # Corresponding data points

# Creating a figure and an axes
fig, ax = plt.subplots()

# Plotting the data
ax.plot(x, y, marker='o')  # Using markers for each data point

# Setting labels for axes
ax.set_xlabel('Time (hours)')  # Labeling the x-axis
ax.set_ylabel('Metric (units)')  # Labeling the y-axis

# Hiding the top and right spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Hiding the ticks and labels on the top and right sides
ax.yaxis.set_ticks_position('left')
ax.xaxis.set_ticks_position('bottom')

# Hiding tick labels on both axes
ax.set_xticklabels([])
ax.set_yticklabels([])

# Adding text annotations for specific times
# ax.text(0.0, 0, 'Time = 0', ha='center', va='center')  # 'ha' and 'va' adjust horizontal and vertical alignment
# ax.text(max(x) + 0.0, 0, 'Time = T', ha='center', va='top')  # Near the end, slightly offset to the right
ax.text(0.5, -0.02, 'Time = 0', ha='center', va='top', transform=ax.get_xaxis_transform())
ax.text(5, -0.02, 'Time = T', ha='center', va='top', transform=ax.get_xaxis_transform())


# Title of the plot
ax.set_title('Driving Session Analysis')

# Adjusting the x-axis limits to accommodate text
ax.set_xlim([min(x) - 0.5, max(x) + 0.5])

# Showing the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams.update({
    "font.family": "sans-serif",
    "font.sans-serif": ["Nexus Sans Pro", "Arial"],  # fallback to Arial if Nexus Sans Pro is unavailable
    "font.size": 12
})


# Number of days and hours in a day
num_days = 10
hours_in_day = 24

# Generate an array of hours for 10 days, cycling every 24 hours
hours = np.tile(np.arange(hours_in_day), num_days)

# Prepare cyclical embeddings
sin_hours = np.sin(2 * np.pi * hours / hours_in_day)
cos_hours = np.cos(2 * np.pi * hours / hours_in_day)

# Define the figure and axes
fig, axs = plt.subplots(1, 2, figsize=(12, 5), dpi=300)  # Adjusted for side-by-side subplots

# Plot 1: Cyclic plot of hours over 10 days
axs[0].plot(hours, 'o-', color='tab:blue', markersize=5, linewidth=1, markerfacecolor='tab:blue', markeredgecolor='black', markeredgewidth=0.5)
axs[0].set_xlabel('Hour Index (from 0 to 239)', fontsize=12)
axs[0].set_ylabel('Hour of the Day (0-23)', fontsize=12)
axs[0].set_title('Hourly Plot Over 10 Days', fontsize=14)
axs[0].set_xticks(np.arange(0, num_days * hours_in_day, hours_in_day))  # Set x-ticks to be every 24 hours
axs[0].set_xticklabels(['Day 1', 'Day 2', 'Day 3', 'Day 4', 'Day 5', 'Day 6', 'Day 7', 'Day 8', 'Day 9', 'Day 10'], fontsize=10, rotation=45)
axs[0].set_yticks(np.arange(0, 24, 3))  # Y-ticks every 3 hours
axs[0].grid(True)

# Plot 2: Cyclical Embedding
scatter = axs[1].scatter(sin_hours, cos_hours, c=hours, cmap='viridis', edgecolors='k', linewidths=0.5)
axs[1].set_xlabel('sin(2π * hour/24)', fontsize=12)
axs[1].set_ylabel('cos(2π * hour/24)', fontsize=12)
axs[1].set_title('Cyclical Embedding of Hours', fontsize=14)
axs[1].set_aspect('equal', 'box')  # Set equal scaling by changing aspect ratio

# Color bar for the scatter plot
cbar = plt.colorbar(scatter, ax=axs[1], orientation='vertical')
cbar.set_label('Hour of Day', fontsize=12)

# Improve layout
plt.tight_layout()

# Save the figure as a PDF file
plt.savefig('hourly_data_plots_professional.pdf', format='pdf', bbox_inches='tight')

# Display the plot
plt.show()
```
