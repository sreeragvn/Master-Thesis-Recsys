```python
import pandas as pd
import os
```

```python
path_merged = '../data/04_Merged'
vehicles = ['SEB880','SEB882','SEB883','SEB885','SEB888','SEB889']
```

```python
dfs = []
for vehicle in vehicles:
    df = pd.read_csv(os.path.join(path_merged, vehicle+'_merged.csv'), low_memory=False)
    df['car'] = vehicle
    dfs.append(df)
data = pd.concat(dfs, axis=0)
```

```python
parsed_esotrace_path = '../data/01_Eso_Extracted'
vehicles = ['SEB880','SEB882','SEB883','SEB885','SEB888','SEB889']

dfs = []
for vehicle in vehicles:
    df = pd.read_csv(os.path.join(parsed_esotrace_path, vehicle+'_extracted_eso.csv'), low_memory=False)
    df['car'] = vehicle
    dfs.append(df)
data = pd.concat(dfs, axis=0)
```
